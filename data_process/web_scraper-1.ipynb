```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from selenium import webdriver
```

```python
driver = webdriver.Chrome("/Users/bennysun/Downloads/chromedriver")
driver.get("https://www.imdb.com/search/title/?title_type=feature&year=2023-01-01,2023-12-31&ref_=adv_prv")
```

```python
headers = driver.find_elements_by_xpath('//h3[@class="lister-item-header"]')
```

```python
title_list = []
for head in headers:
    val = ' '.join(head.text.split(" (2023)")[0].split(". ")[1:])
    title_list.append(val)
```

```python
genre = driver.find_elements_by_xpath('//span[@class="genre"]')
```

```python
genre_list = []
for g in genre:
    genre_list.append(g.text)
```

```python
desc = driver.find_elements_by_xpath('//p[@class="text-muted"]')
```

```python
desc_list = []
for d in desc:
    desc_list.append(d.text)
```

```python
elements = driver.find_elements_by_xpath('//p[@class]')
```

```python
crew_list = []
for d in elements:
    text = d.text
    director = "Director"
    if text.startswith(director):
        crew_list.append(text)
```

```python
v = crew_list[0].split(' | ')
v[1].split(": ")[1].split(", ")
```

```python
direct_list = []
act_list = []
for val in crew_list:
    txt = val.split(' | ')
    direct_list.append(','.join(txt[0].split(": ")[1].split(", ")))
    if len(txt) == 2:
        act_list.append(','.join(txt[1].split(": ")[1].split(", ")))
    else:
        act_list.append("")
```

```python
data = [title_list, genre_list, desc_list, direct_list, act_list]
df = pd.DataFrame(data).T
df.columns = ["title", "genre", "desc", "director", "actors"]
```

```python
unique_id = []
for val in df.index.values.astype(str):
    unique_id.append("id" + val)
```

```python
df["nmid"] = unique_id
```

```python
cols = ["nmid", "title", "genre", "desc", "director", "actors"]
```

```python
df = df[cols]
```

```python
import requests
for __, row in df.iterrows():
    
    val = row["actors"] if row["actors"] != "" else " "
    
    body = {
        "nmid":row["nmid"],
        "title":row["title"],
        "genre":row["genre"],
        "desc":row["desc"],
        "directors":row["director"],
        "actors":val
    }
    url = "http://localhost:5001/betmovie"
    requests.put(url, body)
```
